```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
data = pd.read_csv('Indicators.csv')
```

```python
#dataset from https://www.kaggle.com/worldbank/world-development-indicators
```

```python
life = data['IndicatorName'].str.contains('Life expectancy') & data['IndicatorName'].str.contains('total')
ld1 = data[life]
ld1.head(20)
```

```python
ld1.isnull().any()
```

```python
ldr1 = ld1[(ld1.Year > 2010)]
ldr1.describe()
```

```python
ldr1_75 = ldr1[(ldr1.Value > 76.8)]
ldr1_75.head(20)
```

```python
sorted = ldr1_75.sort_values(by='Value', ascending=False)
```

```python
#remove non-country region designations
regions = ['EMU', 'EUU', 'HIC', 'OEC', 'OED', 'NAC']
sorted = sorted[-sorted.CountryCode.isin(regions)]
sorted.head(20)
```

```python
sorted2013 = sorted[sorted.Year == 2013]
sorted2013.head(20)
```

```python
top20 = sorted2013.head(20).CountryCode.values.tolist()
print(top20)
print(len(top20))
```

```python
#for top 20 countries does health spending correlate with life expectancy
spend = data['IndicatorName'].str.contains('Health expenditure') & data['IndicatorName'].str.contains('PPP')
s1 = data[spend]
s1.head()
```

```python
#healthcare expenditures for 2013
s1[s1.Year == 2013].describe()
```

```python
s1_20 = s1[s1.CountryCode.isin(top20)]
s1_20.head()
s1_20.count()
```

```python
s1_2013 = s1_20[s1_20.Year == 2013]
s1_2013.head()
s1_2013.count()
```

```python

joined = s1_2013.merge(sorted2013, on='CountryCode', how='left')
joined.head()
```

```python
del joined['IndicatorCode_x']
```

```python
del joined['IndicatorCode_y']
```

```python
del joined['Year_y']
```

```python
del joined['CountryName_y']
joined.head(20)
```

```python
joined = joined.sort_values(by='Value_y', ascending=False)
joined.head(20)

```

```python
joined.Value_x.max()
#Luxembourg
```

```python
joined.Value_x.min()
#Chile
```

```python
df = joined.set_index('CountryName_x')
ax = df['Value_x'].plot(kind='bar', title ="2013 Health Expenditures - Top Life Expectancy Countries", 
                        figsize=(15, 10), legend=True, fontsize=12)
ax.set_ylabel("Expenditure Per Capita", fontsize=12)
plt.show()
```

```python
colors = np.random.rand(3,1)
plt.scatter(joined.Value_y, joined.Value_x, c=colors, s=200)

for label, x, y in zip(joined.CountryName_y, joined.Value_y, joined.Value_x):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-20, 20),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

plt.show()
```
